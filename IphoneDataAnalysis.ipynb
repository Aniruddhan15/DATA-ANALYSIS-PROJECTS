```python
plt.figure(figsize=(17,10))
sns.barplot(x='Product Name',y='Star Rating',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
df['Product Name'].nunique()
```

```python
plt.figure(figsize=(17,10))
sns.barplot(x='Product Name',y='Number Of Reviews',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(17,10))
sns.barplot(x='Product Name',y='Star Rating',hue='Ram',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(17,10))
sns.barplot(x='Ram',y='Number Of Ratings',data=df)
plt.xticks(rotation=90)
plt.show()
```
